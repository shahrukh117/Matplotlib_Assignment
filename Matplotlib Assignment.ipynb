### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a popular Python library for creating static, animated, or interactive visualizations in a variety of formats, such as charts, plots, and graphs. It provides a wide range of functionality for creating high-quality visualizations and is extensively used in data analysis, scientific research, and data visualization tasks.

Five types of plots that can be created using the Pyplot module of Matplotlib include:

#### Line Plot:

A line plot, or line chart, is used to display data points as connected line segments. It is often used to visualize trends and changes in data over time or across a continuous variable.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

#### Scatter Plot:

A scatter plot is used to represent individual data points as markers on a 2D plane. It is suitable for visualizing the relationship between two variables, identifying outliers, and clustering.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

#### Bar Plot:

A bar plot is used to represent categorical data with rectangular bars. It is effective for comparing quantities across different categories or groups

```python
import matplotlib.pyplot as plt

categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [15, 25, 20, 30]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

#### Histogram:

A histogram is used to represent the distribution of a single variable by dividing it into bins and showing the frequency or count of data points in each bin. It's useful for understanding data distributions.

```python
import matplotlib.pyplot as plt

data = [3, 4, 4, 5, 5, 5, 6, 6, 7, 7, 7, 8, 8, 8, 8]

plt.hist(data, bins=5)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

#### Pie Chart:

A pie chart is a circular graph that represents the parts of a whole. It's useful for showing the composition or distribution of a dataset, especially when dealing with categorical data.

```python
import matplotlib.pyplot as plt

labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 30, 15, 30]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

##### Scatter Plot:
A scatter plot is used to represent individual data points as markers on a 2D plane. It is suitable for visualizing the relationship between two variables, identifying outliers, and clustering.

```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(3)


x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y, color='blue', marker='o', label='Data Points')

# Add title and axis labels
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
#### Use the following data:
import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The subplot() function in Matplotlib is used to create multiple plots in a single figure. It is particularly useful when you want to display multiple graphs or plots side by side or in a grid within the same figure. This can be helpful for comparing and visualizing multiple datasets or different aspects of the same data

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for four line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)  # Create a 2x2 grid of plots, select the first plot
plt.plot(x, y1)
plt.title('Line 1')

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)  # Select the second plot
plt.plot(x, y2)
plt.title('Line 2')

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)  # Select the third plot
plt.plot(x, y3)
plt.title('Line 3')

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)  # Select the fourth plot
plt.plot(x, y4)
plt.title('Line 4')

# Adjust layout for better visualization
plt.tight_layout()

# Show the plots
plt.show()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

A bar plot, also known as a bar chart or bar graph, is a graphical representation of data in which rectangular bars of varying heights or lengths are used to show the values of different categories or data points. The length or height of each bar is proportional to the value it represents, making it a useful tool for comparing and visualizing the data of different categories.

Bar plots are commonly used for the following purposes:

Comparison: Bar plots make it easy to compare the values of different categories or data points. You can quickly see which category has the highest or lowest value.

Categorical Data: Bar plots are particularly useful for displaying categorical data, such as comparing sales of different products, the performance of different companies, or any data where categories can be clearly defined.

Trends and Patterns: Bar plots can also be used to visualize trends or patterns over time or across categories.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical Bar Plot
plt.figure(figsize=(8, 4))  # Adjust the figure size
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.title('Company Profits (Vertical Bar Plot)')

plt.show()

# Horizontal Bar Plot
plt.figure(figsize=(8, 4))  # Adjust the figure size
plt.barh(company, profit, color='lightcoral')
plt.xlabel('Profit (in billions)')
plt.ylabel('Company')
plt.title('Company Profits (Horizontal Bar Plot)')

plt.show()
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It provides a summary of the data's central tendency, spread, and the presence of outliers. Box plots are commonly used for the following purposes:

Data Distribution: Box plots display the central tendency of the data (median), the spread of the data (interquartile range), and any potential outliers. They give you a quick sense of the overall distribution.

Comparison: Box plots are useful for comparing the distribution of multiple datasets or groups. They help you understand how the data in different categories or groups compare in terms of their central tendency and variability.

Identifying Outliers: Box plots can help identify potential outliers in the data, which are data points that significantly differ from the rest of the dataset.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

# Create a box plot
plt.figure(figsize=(8, 6))  # Adjust the figure size
plt.boxplot(data, labels=['Box 1', 'Box 2'], vert=False)
plt.xlabel('Value')
plt.ylabel('Box')
plt.title('Box Plot of Two Datasets')

plt.show()
```
